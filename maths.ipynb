```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
x = np.linspace(0, 2*np.pi, 1000)
y = np.sin(x)

# Plot the data as a line
plt.plot(x, y, '-')

# Add axis labels and title
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Sine Wave')

# Show the plot
plt.show()

```
