```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_excel("C:\\Users\\Hp\\OneDrive\\Desktop\\Excel\\Rent.xlsx")
df
```

```python
plt.xlabel('area(sqr ft)')
plt.ylabel('rent(us$)')
plt.scatter(df.area,df.rent,color = 'red')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.rent)
input_data = np.array([input("enter desired area of house")],dtype=float)
reshaped_data = input_data.reshape(1,-1)
predection = reg.predict(reshaped_data)
print(predection)
```
